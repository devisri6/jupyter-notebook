```python
region_count = df["Regionname"].value_counts()
region_count = dict(region_count)
a = region_count.keys()
b = region_count.values()
```

```python
#pie chart
fig = plt.figure(figsize=(8, 8))

plt.pie(x = b ,labels = a)
plt.legend()
plt.show()
```

```python
#histogram
region = df["Regionname"]
fig = plt.figure()
plt.hist(x=region, bins=10)
plt.hist(x=region, edgecolor='black')
plt.show
```

```python
type_count = df['Type'].value_counts()
type_count = dict(type_count)
```

```python
print(type_count)
```

```python
k = list(type_count.keys())
v = list(type_count.values())
```

```python
fig = plt.figure(figsize=(4,4))

plt.bar(x = k, height = v) #bar plotting syntax for bar plot is plt.bar
```

```python
#line plot
price = df["Price"]
price = list(price)
distance = df["Distance"]
distance = list(distance)
```

```python
fig = plt.figure(figsize = (8,8))
#line plot
plt.plot(distance,price, linestyle = "", marker = "*", markerfacecolor = 'bluemarkeredgecolor = 'black')
plt.xlabel("distance")
plt.ylabel("price")
plt.title("distance vs price")
plt.show()
```
